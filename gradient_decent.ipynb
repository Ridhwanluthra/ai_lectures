# Gradient Decent

### Task
Function J($\theta$1, $\theta$2)

Goal: Min of J w.r.t. $\theta$1, $\theta$2

### Update rule:

$\theta$j = $\theta$j - $\alpha$$\frac{\partial v}{\partial t}$J($\theta$1, $\theta$2)

$\alpha$ = learning rate

## Let's impliment our own gradient decent

#### Declare Some Variables

```python
# Staring point
cur_x = 3
# Learning rate
rate = 0.01
# This tells us when to stop the algorithm
precision = 0.000001
previous_step_size = 1
# maximum number of iterations
max_iters = 10000
#iteration counter
iters = 0
```

#### Define our function and it's gradient

```python
# Our function
f = lambda x: (x+5)**2
# Gradient of our function
df = lambda x: 2*(x+5)
```

```python
# f = lambda x: x**4  - 20*x**2 + 10*x
# df = lambda x: 4*x**3 - 40*x + 10
```

### Gradient Decent Algorithm

```python
while previous_step_size > precision and iters < max_iters:
    # Store current x value in prev_x
    prev_x = cur_x
    # Grad descent step
    cur_x = cur_x - rate * df(prev_x)
    # Change in x
    previous_step_size = abs(cur_x - prev_x)
    #iteration count
    iters += 1
    print("Iteration",iters,"\nX value is",cur_x) #Print iterations
    
print("The local minimum occurs at", cur_x)
```
